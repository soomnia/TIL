## textwrap
- 파이썬에서 **문자열을 보기 좋은 형태**로 정렬 또는 줄바꿈하는데 사용할 수 있는 라이브러리
- 원하는 길이에 맞게 줄이기 : textwrap.shorten(문자열, width=n, placeholder='...[이하줄임]')
- 긴 문장 자르기 : textwrap.wrap(문자열, width=n)
- 긴 문장 줄바꿈 : textwrap.fill(문자열, width=n)

### textwrap.shorten(문자열, width=글자수, placeholder='말줄임표')

```python
textwrap.shorten(text, width=200)
```

```python
textwrap.shorten(text, width=100, placeholder='...[중략]')
```

### textwrap.wrap(문자열, width=글자수)
width 크기만큼씩 문장을 나눔

```python
wrapped_text = textwrap.wrap(text, width=40)
print(wrapped_text)
```

```python
# 줄바꿈을 기준으로 join하여 보기 좋게 print
print('\n'.join(wrapped_text))
```

### textwrap.fill(문자열, width=글자수)
위의 join이 내포된 기능

```python
filled_text = textwrap.fill(text, width=40)
print(filled_text)
```

## 정규표현식 re
`regular expression`<br>

복잡한 문자열 처리 시 사용. 특정 문자열 **추출**, **변환** 등에 사용 <br>

[공식문서](https://docs.python.org/3/library/re.html)

### re.findall(규칙, 문자열)
단어 추출

```python
import re
```

```python
# 단어 추출
## r'\w+': 단어 단위로 추출하라는 정규표현식
words = re.findall(r'\w+', text)
print(words)
```

```python
## 전화 번호 추출
# 연락처 목록
contact = '''김미키 21 010-3344-5566 Mike@google.com 
             김소은 20 010-5032-1111 Soeun@naver.com
             유한슬 34 010-2789-1476 Lyu@school.ac.com
             박민철 40 010 4040 1313 Zoe@school.ac.com
             이민아 23 010-7777-2222 Kate@google.com'''

# 전화번호 추출
## d: diget
regex = r'0\d{1,2}[ -]?\d{3,4}[ -]?\d{3,4}'

phone = re.findall(regex, contact)
print("\n".join(phone))
```

```python
# 문자열 변환(마스킹)
pat = re.compile(r'0\d{1,2}[ -]?\d{3,4}[ -]?\d{3,4}')
## sub: substitute
print(pat.sub("***-****-****", contact))
```
