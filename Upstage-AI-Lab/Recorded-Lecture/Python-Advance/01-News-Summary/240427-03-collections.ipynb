```python
# 문자열 편집
text = textwrap.fill(text, width=50)
print(text)
```

```python
# 단어 추출
words = re.findall(r'\w+', text)
print(words)
```

## Counter(단어리스트)
딕셔너리 형태. 덧셈 등의 연산 가능

```python
# 빈도수 산출
counter = collections.Counter(words)
print(counter)
```

```python
# 덧셈
a = collections.Counter(['a', 'b', 'c', 'b', 'd', 'a'])
b = collections.Counter(['e', 'f', 'f', 'b', 'a', 'd'])
```

```python
print(a)
print(b)
print(a+b)
```

```python
# 뺄셈
print(a-b)
```

```python
# 합집합
print(a | b)
```

```python
# 교집합
print(a & b)
```
