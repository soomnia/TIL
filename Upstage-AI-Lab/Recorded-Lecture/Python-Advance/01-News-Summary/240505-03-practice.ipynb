## 1. 디코딩
바이너리 파일 읽기

```python
# 기사 이미지 읽기
## rb: read bytes
b_img = open("news/image", 'rb')
img = b_img.readlines()
b_img.close()
```

```python
# 기사 본문 읽기
f = open("news/article", 'rb')
article = f.readlines()
f.close()
```

```python
# 디코딩
import base64
```

```python
img_encoded = img[0]

path = 'news/image.jpg'

with open(path, 'wb') as f:
    img_decoded = base64.decodebytes(img_encoded)
    f.write(img_decoded)
```

```python
# 기사 디코딩
article_encoded = article[0]
article_decoded = base64.decodebytes(article_encoded)
# bytes 형태의 string으로 디코딩됨
article_decoded
```

```python
# utf-8로 한 번 더 디코딩
article = article_decoded.decode('utf-8')
print(article)
```

## 2. 문서 요약 + 키워드 추출

```python
# 내용 요약
from gensim.summarization.summarizer import summarize
#
from gensim.summarization.textcleaner import split_sentences
```

```python
# 단어수 기반 요약 >> word_count
summarize(article, word_count=50)
```

```python
# 비율 기반 요약 >> ratio
summarize(article, ratio=0.1)
```

```python
summarized = summarize(article, ratio=0.1)
```

```python
# 키워드 추출
import collections
import textwrap
import re
```

```python
# 줄바꿈 정렬
article_align = textwrap.fill(article, width=50)
article_align
```

```python
# 단어 추출
words = re.findall(r'\w+', article_align)
words
```

```python
# 빈도수 산출
counter = collections.Counter(words)
counter
```

```python
# 키워드 추출
counter.most_common(5)
```

```python
# '있다'는 크게 의미 있진 않으므로 제외하고
keywords = counter.most_common(5)[1:]
keywords
```

## 3. 요약 리포트 작성

```python
from IPython.core.display import Image
Image(filename=path, width=300)
```

```python
hashtags = ['#' + keyword[0] for keyword in keywords]
hashtags = ' '.join(hashtags)
hashtags
```

## 4. html 파일로 저장

```python
with open('news/summary.html', 'w') as html_file:
    html_file.write("<html>\n")
    html_file.write("<h1>카운트다운 들어간 아르테미스 계획... \"달의 여신\"은 미소 지을까?</h1>\n")
    html_file.write("<img src='image.jpg'/>\n")
    html_file.write(f"<h2>{summarized}</h2>\n")
    html_file.write(f"<h2 style='background-color:powderblue;'>{hashtags}</h2>\n")
    html_file.write("</html>\n")
```
