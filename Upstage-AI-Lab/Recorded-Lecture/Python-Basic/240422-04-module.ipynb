### from
모듈 내에서 필요한 것만 선택해 활용하고자 할 때 사용
```python
from 모듈_이름 import 가져오고_싶은_변수_또는_함수

# 모듈 모두를 가져오고 싶을 때
from 모듈_이름 import *
```

### as
모듈에 별칭을 붙일 때 사용
* 모듈을 가져올 때 이름이 충돌하는 경우
* 이름이 길어서 줄여 쓰고 싶은 경우

#### seaborn - swarmplot()
매개변수
- x축 변수
- y축 변수
- 데이터 셋
- hue : 특정 열 데이터로 색상을 구분하여 출력

````python
## 데이터 시각화
### 데이터 분산 고려, 데이터 포인트가 서로 중복되지 않도록 시각화 > 입체적 파악 가능
sns.swarmplot(x='class', y='age', data=titanic, hue='sex')
````
