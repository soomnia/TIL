## 입력 input
프롬프트 문자열: input() 함수 괄호 안에 입력한 내용(사용자로부터 입력을 요구하는 내용)
- input() 함수의 결과 값을 변수에 대입
- input() 함수로 받은 입력 결과는 무조건 <b>문자열</b> <br>
따라서 적절한 형변환을 통해 사용해야 함

````python
str1 = input("1을 입력하세요")
str2 = input("2을 입력하세요")
print("입력값 1의 type은", type(str1))
print("입력값 2의 type은", type(str2))

# 형변환 - int()
int1 = int(str1)
int2 = int(str2)

# 문자열 덧셈
print("문자열 덧셈:", str1+str2)

# 정수 덧셈
print("정수 덧셈:", int1+int2)

# 형변환 - float()
float1 = float(str1)
float2 = float(str2)

# 실수 덧셈
print("실수 덧셈:", float1+float2)

# 형변환 - 문자열
int3 = 123
print(int3, "의 타입은", type(int3))

str3 = str(int3)
print(str3, "의 타입은", type(str3))
````

### 읽기
파일을 읽기 모드로 열기
* readline()<br>
한 줄씩 읽어서 출력. 줄 끝에 \n이 있다면 빈 줄도 같이 출력됨
* readlines()<br>
파일의 모든 줄을 읽어 각각의 줄을 리스트로 반환
* strip<br>
줄 바꿈(\n) 제거

````python
read_file = open("python.txt", "r")
````

````python
# readline() : 텍스트 한 줄씩 읽어서 반환
# 한 줄 씩 읽어 출력할 때 줄 끝에 \n 문자가 있다면 빈 줄도 같이 출력됨
# 출력할 때마다 한 줄 씩 내려가서 출력됨
# 한 번 실행하면 "Programming! \n", 두 번 실행하면 "Hello" 출력
text = read_file.readline()
print(text)
````

````python
read_file = open("python.txt", "r")
````

### 논리 연산자
* not (논리 부정)<br>
한 개의 값이 False -> True<br>
한 개의 값이 True -> False

* and (논리 곱)<br>
두 개의 값이 모두 True -> True<br>
나머지는 모두 False 반환

* or (논리 합)<br>
두 개의 값이 모두 False -> False<br>
나머지는 모두 True 반환

## for 반복문
### 딕셔너리 반복
- 딕셔너리는 키를 기반으로 여러 자료를 저장하는 자료형
- 키(key)는 딕셔너리 안에서 값을 접근할 때 사용하는 것
- 값(value)은 딕셔너리 안에서 각각의 내용을 의미
- 딕셔너리를 이용한 for문은 딕셔너리 안에 있는 키(key)가 변수에 들어감 

````python
myDictionary = {
    "name": "soom",
    "job": "data scientist"
}
````

````python
for key in myDictionary:
    print(key, ":", myDictionary[key])
````

### 범위 반복
#### range()
- range(A)<br>
매개변수로 숫자 한 개 입력. 0부터 A-1까지의 정수로 범위 생성
- range(A, B)<br>
매개변수로 숫자 두 개 입력. A부터 B-1까지의 정수로 범위 생성
- range(A, B, C)<br>
매개변수로 숫자 세 개 입력. A부터 B-1까지 C만큼 차이 나는 범위 생성

#### list()
범위를 리스트로 변경하면 범위 안에 어떤 값이 있는지 확인 가능

````python
print(list(range(5)))
print(list(range(5, 10)))
print(list(range(5, 10, 2)))
````

## while 반복문
조건을 만족하는 한, 반복적으로 문장 실행 > 무한 반복이 되지 않도록 주의

### break
반복문을 벗어날 때 사용하는 키워드<br>
일반적으로 무한 반복문을 만들고 그 내부를 벗어나고자 할 때 사용

````python
a = 10
count = 0

while a > 3:
    count += 1
    a += 1
    if count > 20:
        break
    print("a :", a, "\tcount :", count)

print("반복문 종료")
````

### continue
반복문의 현재 진행 과정을 생략하고 다음 반복으로 넘어가고자 할 때 사용

````python
num = 0
while num < 10:
    num += 1
    if num <= 5:
        continue
    print(num)
````
